```python
import os
from matplotlib.image import imread
import matplotlib.pyplot as plt

import numpy as np
import os

# important step to locate and give access to drive
from google.colab import drive
drive.mount('/content/drive')
os.chdir('/content/drive/My Drive/Classroom/DataDrivenModeling/ClassNotes/L1to3_SVD/codes/')

plt.rcParams['figure.figsize'] = [16, 8]

#A = imread(os.path.join('../..','DATA','dog.jpg'))
A = imread('dog.jpg')

X = np.mean(A, -1); # Convert RGB to grayscale

# original
img0=plt.imshow(A)
#img0.set_cmap('hsv') # try this option and mis-spell hsv as hasv
plt.axis('off')   # try by commenting this line
plt.show()

# gray
img = plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U, S, VT = np.linalg.svd(X,full_matrices=False)
S = np.diag(S)

j = 0
for r in (5, 20, 100):
    # Construct approximate image
    Xapprox = U[:,:r] @ S[0:r,:r] @ VT[:r,:]
    plt.figure(j+1)
    j += 1
    img = plt.imshow(Xapprox)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('r = ' + str(r))
    plt.show()

## Let us plot their singular values
plt.figure(1)
plt.semilogy(np.diag(S))
plt.title('Singular Values')
plt.show()

plt.figure(2)
plt.plot(np.cumsum(np.diag(S))/np.sum(np.diag(S)))
plt.title('Singular Values: Cumulative Sum')
plt.show()
```
