```python
import os
from matplotlib.image import imread
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
from PIL import Image
```

```python
myImage = Image.open("/Users/venkatasaisarangrandhe/Downloads/IMG_1730.jpg");
myImage.show();
```

```python
img0 = plt.imshow(myImage)
plt.axis('off')
plt.show()
```

```python
X = np.mean(myImage, -1)
img = plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U, S, VT = np.linalg.svd(X, full_matrices = False)
S = np.diag(S)
```

```python
j = 0
for r in (5,20,100):
    Xapprox = U[:,:r]@S[0:r,:r]@VT[:r,:]
    plt.figure(j+1)
    
    j+=1
    img = plt.imshow(Xapprox)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('r = ' + str(r))
    plt.show()
```

```python
plt.figure(1)
plt.semilogy(np.diag(S))
plt.title('Singular Values')
plt.show()

plt.figure(2)
plt.plot(np.cumsum(np.diag(S))/np.sum(np.diag(S)))
plt.title('Singular Values: Cumulative Sum')
plt.show()
```
