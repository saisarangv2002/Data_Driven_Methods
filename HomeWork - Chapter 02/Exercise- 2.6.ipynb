```python
#(a) Left boundary = 1 and right boundary = 2, zero initial conditions
grid_size = 300

# Define the range for x and y
x_min, x_max = -1.5, 1.5
y_min, y_max = -1.5, 1.5
x_range = np.linspace(x_min, x_max, grid_size)
y_range = np.linspace(y_min, y_max, grid_size)

# Create a grid using numpy.meshgrid
x_grid, y_grid = np.meshgrid(x_range, y_range)

# Calculate the distance from the origin for each point
distance_grid = np.sqrt(x_grid**2 + y_grid**2)

u = np.zeros((grid_size, grid_size))
#grid_with_zeros[distance_grid > 1] = 0
```

```python
u[(distance_grid>1)*(x_grid<0)] = 1
u[(distance_grid>1)*(x_grid>=0)] = 2
plt.imshow(u)
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
for iteration in range(1000):
    #if iteration%10==0:
    lu = u       
    for i in range(1,299):
        for j in range(1,299):
            u[i,j]= (1/4)*(u[i+1,j]+u[i-1,j]+u[i,j+1]+u[i,j-1]) #This is the laplacian of the variable u
    u = lu
    if iteration==100:
        plt.imshow(lu)
        plt.colorbar()
        plt.show()
    u[(distance_grid>1)*(x_grid<0)] = 1
    u[(distance_grid>1)*(x_grid>=0)] = 2

plt.imshow(lu)
plt.colorbar()
plt.show()
```
