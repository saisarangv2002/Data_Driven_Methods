```python
import math
a = [0]*100
b = [0]*100
a[0] = 1/2
for n in range(1,100):
    a[n]= -2*((4*np.cos(n*math.pi/2)-4)/(math.pi**2*n**2))

plt.plot(range(0,100),a)
plt.plot(range(0,100),b)
plt.yscale("log")
plt.xlabel("n value")
plt.ylabel("a[n]")
plt.title("a[n], b[n] vs n")
plt.show()
```

```python
dx = 0.001
L = np.pi
x = L * np.arange(-1+dx,1+dx,dx)
n = len(x)
nquart = int(np.floor(n/4))
# Define hat function
f = np.zeros_like(x)
f[nquart:2*nquart] = (4/n)*np.arange(1,nquart+1)
f[2*nquart:3*nquart] = np.ones(nquart) - (4/n)*np.arange(0,
   nquart)
ffs =  a[0]/2
for k in range(1,10):
    ffs = ffs + a[k]*np.cos(k*math.pi*x/2)
plt.plot(x, f)
plt.plot(x, ffs)
plt.show()
```
