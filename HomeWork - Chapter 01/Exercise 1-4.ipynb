# Exercise 1-4. Generate a random 100×100 matrix, i.e., a matrix whose entries are sampled from a normal distribution. Compute the SVD of this matrix and plot the singular values. Repeat this 100 times and plot the distribution of singular values in a box-and-whisker plot. Plot the mean and median singular values as a function of r. Now repeat this for different matrix sizes (e.g., 50×50, 200×200, 500 × 500, 1000 × 1000, etc.).

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [40,15]

# Function to compute SVD and return singular values
def compute_svd():
    A = np.random.randn(100, 100)
    U, S, Vh = np.linalg.svd(A, full_matrices=False)
    return S

# Number of iterations
num_iterations = 100

# List to store singular values from each iteration
singular_values = []

# Perform SVD 100 times and store singular values
for i in range(num_iterations):
    singular_values.append(compute_svd())

# Convert the list of singular values to a NumPy array for easy manipulation
singular_values = np.array(singular_values)

# Plot the singular values
plt.stem(singular_values[0])
plt.xlabel("Rank")
plt.ylabel("Singular value")
plt.title("Singular values of a random 100x100 matrix")
plt.show()

# # Plot the distribution of singular values in a box-and-whisker plot
# plt.boxplot(singular_values)
# plt.xlabel("Rank")
# plt.ylabel("Singular value")
# plt.title("Distribution of singular values of 100 random 100x100 matrices")
# plt.show()

# Plot the distribution of singular values in a box-and-whisker plot
plt.boxplot(singular_values.T)  # Transpose the singular_values array
plt.xlabel("Rank")
plt.ylabel("Singular value")
plt.title("Distribution of singular values of 100 random 100x100 matrices")
plt.show()

# Compute the mean and median singular values
mean_singular_values = np.mean(singular_values, axis=0)
median_singular_values = np.median(singular_values, axis=0)

# Plot the mean and median singular values as a function of rank
plt.plot(mean_singular_values, label="Mean")
plt.plot(median_singular_values, label="Median")
plt.xlabel("Rank")
plt.ylabel("Singular value")
plt.title("Mean and median singular values of 100 random 100x100 matrices")
plt.legend()
plt.show()
```
